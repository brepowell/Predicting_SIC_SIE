```python
import matplotlib.pyplot as plt
import numpy as np

# Step names
steps = [
    "Data/Time Loading",
    "Freeboard Calculation",
    "Patchifying",
    "Plotting Outliers/Imbalance",
    "Freeboard Normalization"
]

# Elapsed times in seconds
full_times = [
    1234.8578193187714,
    1307.4704449176788,
    1308.1274116039276,
    2314.5647683143616,
    2332.2917351722717
]

# Replace these with your trial version's actual timings
trial_times = [
    136.7,    # e.g., time for loading raw data in trial init
    152.1,    # freeboard calc
    149.4,    # patchifying
    195.8,    # outlier analysis
    209.3     # normalization
]

# Set up bar positions
x = np.arange(len(steps))
width = 0.35

# Plot
plt.figure(figsize=(12, 6))
plt.bar(x - width/2, trial_times, width, label='Trial Dataset', color='skyblue')
plt.bar(x + width/2, full_times, width, label='Full Dataset', color='salmon')

# Labels and formatting
plt.ylabel('Elapsed Time (seconds)')
plt.title('Initialization Step Timings: Trial vs Full Dataset')
plt.xticks(x, steps, rotation=20)
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Annotate bars
for i in range(len(steps)):
    plt.text(x[i] - width/2, trial_times[i] + 20, f"{trial_times[i]:.1f}s", ha='center', fontsize=9)
    plt.text(x[i] + width/2, full_times[i] + 20, f"{full_times[i]:.1f}s", ha='center', fontsize=9)

plt.tight_layout()
plt.savefig("Dataset init timings - trial and full comparison.png")
plt.show()
plt.close()
```
